interpretability testing

```python
MODEL_NAME       = 'roberta-large-mnli'
DEVICE           = 'auto'
THRESHOLD        = 0.5

target_text = "[PRED] Died [/PRED] : [ARG] James A. Attwood , 62 , retired chairman and president of Mutual Life Insurance Co. of New York [/ARG] , Tuesday , in New York City , of an acute anemic condition ."

hypothesis = "James A. Attwood is a sentient being."
```

```python
from probing import prober
from interpretability import shapley

test_prober = prober.Prober(MODEL_NAME, device=DEVICE, threshold=THRESHOLD)
shapley_prober = shapley.ShapleyProber(test_prober)
```

```python
test_pair = [[target_text, hypothesis]]

test_background = [["[ARG] Vincent [/ARG] violated [PRED] the dress code [/PRED].", "Vincent wore improper clothing."]]

#shap_values = shapley_prober.explain(test_pair, test_background, visualize=True)
shap_values = shapley_prober.kernelexplain(test_pair, test_background)
```

```python
from probing import prober
from interpretability import lime


test_prober = prober.Prober(MODEL_NAME, device=DEVICE, threshold=THRESHOLD)
limeprober = lime.LimeProber(test_prober)

limeprober.explain(target_text, hypothesis)
```
